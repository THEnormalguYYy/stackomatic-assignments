```python
df.describe()# Stasticial summary of data
```

```python
uniquebedrooms = df['total_bedrooms'].unique()#filter distinct numbers in total_bedrooms
no_uniquebedrooms = len(uniquebedrooms)#number of elements in uniquebedrooms

display(uniquebedrooms)
no_uniquebedrooms
```

```python
newdf=df
newdf['total_bedrooms_per_total_rooms']=newdf['total_bedrooms']/newdf['total_rooms']#create new column
print(newdf['total_bedrooms_per_total_rooms'].mean())
print(newdf['total_bedrooms_per_total_rooms'].std())
```
